## PART 4: Total by hospital type

This is a multistep process:
* From our downloaded data file, compute the `Total payments` per ZIP code and Medicaid EP Hospital Type.
* Then merge that with the `population` data to compute a `Total payments` per person.
* Then average that across all of the `Medicaid EP Hospital Types` to get an average per persona payment for these type of hospital.

Your answer should be in structure of a data frame with at least two columns:
* Medicaid_EP_Hospital_Type
* Avg_Pay_per_Capita

## First using SQL

Please see below sql area for pandas solution - much better!!!

```python
#read in the data and see total payments 
pd.read_sql_query("SELECT NPI, Medicaid_EP_Hospital_Type, Business_ZIP_Code,total_payments, SUM(total_payments) AS total FROM lillyeversman_data GROUP BY Business_ZIP_Code, Medicaid_EP_Hospital_Type ", conn)
```

```python
sql = pd.read_sql_query("SELECT NPI, Medicaid_EP_Hospital_Type, Business_ZIP_Code,total_payments, SUM(total_payments) AS total FROM lillyeversman_data GROUP BY Business_ZIP_Code ORDER BY total DESC", conn)
```

```python
#merge the table with popualtion table to calculate the per capital totals
avgcapital = pd.read_sql_query("""
  SELECT NPI, Medicaid_EP_Hospital_Type, Business_ZIP_Code,total_payments, population, SUM(total_payments) AS total
  FROM 
    lillyeversman_data h INNER JOIN
    population v ON h.Business_ZIP_Code = v.Zip
  GROUP BY Business_ZIP_Code ORDER BY total DESC
""", conn)
```

```python
avgcapital['Avg_Pay_per_Capital'] = avgcapital['total']/avgcapital['population']
avgcapital
```

```python
#find averages
import numpy as np
avgcapital2 = pd.pivot_table(data = avgcapital, index = ['Medicaid_EP_Hospital_Type'], values = 'Avg_Pay_per_Capital', aggfunc = np.mean)
avgcapital2
```
